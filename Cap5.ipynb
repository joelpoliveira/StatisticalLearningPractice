```python
def root_mean_sqr_error(model, X, y):
    ols = model.fit(X, y)
    pred = ols.predict(X)
    err = np.sqrt(((pred-y)**2).mean())
    return err
```

```python
def alpha(X, y, index = None):
    if index is not None:
        X = X[index]
        y = y[index]
    return (np.var(y) - np.cov(X, y)[0,1]) / (np.var(X) + np.var(y) - 2*np.cov(X, y)[0,1])
```

```python
def boot(data, func, n):
    alphas = []
    for i in range(n):
        index = data.sample(n = 100, replace = True).index
        alphas.append(func(data.X, data.Y, index))
    alphas = np.array(alphas)
    return alphas.std()/np.sqrt(len(alphas))
```

```python
np.random.seed(2)
x = np.random.normal(size=100)
y = x-2*x**2+np.random.normal(size=100)
```

```python
plt.scatter(x,y)
```

```python
def poly(data, n):
    try:
        len(data)
    except:
        data=[data]
    pol = PolynomialFeatures(degree=n)
    X = pol.fit_transform( np.array(data).reshape(len(data), 1))
    return pd.DataFrame(X)
```

```python
def LOOCV(df, n):
    results = []
    
    for i in df.index:
        index = df.index.difference([i])
        
        train = df.iloc[index].reset_index(drop=True)
        test = df.iloc[i]

        X=poly(train.x, n)
        model = lm.OLS(train.y, X).fit()
        
        test_point = poly(test.x, n)
        pred = model.predict(test_point)
        results.append((test.y - pred)**2 )
        
    return str(np.mean(results))
```

```python
df = pd.DataFrame()
df["y"] = y
df["x"] = x
```

```python
print("LOOCV, order=1: " + LOOCV(df, 1),
      "LOOCV, order=2: " + LOOCV(df, 2),
      "LOOCV, order=3: " + LOOCV(df, 3),
      "LOOCV, order=4: " + LOOCV(df, 4),
      sep="\n")
```

```python
SE = df.medv.std()/np.sqrt(len(df.index))
SE
```

```python
def est_mean(data, index = None):
    return data[index].mean()
```

```python
def boot(data, func, n, k = 100):
    alphas = []
    for i in range(n):
        index = data.sample(n = k, replace = True).index
        alphas.append(func(data, index))
    alphas = np.array(alphas)
    return alphas.std()/np.sqrt(len(alphas))
```

```python
boot(df["medv"], est_mean, n=500, k = 100) 
```
