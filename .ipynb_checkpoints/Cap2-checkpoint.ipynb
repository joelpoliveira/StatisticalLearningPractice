```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import sklearn.datasets as data_sets
```

```python
mat = np.array([[0, 3, 0, 0],
               [2, 0, 0, 0],
               [0, 1, 3, 0],
               [0, 1, 2, 1],
               [-1, 0, 1, 1],
               [1, 1, 1, 0]], dtype = int)
```

```python
 def euclidean_dist(a, b):
        return np.sqrt( np.sum( (a[:-1]-b[:-1])**2 ) )
```

```python
arr = np.array([0,0,0,np.nan])
dist_mat = np.array(list(map(lambda x: euclidean_dist(x, arr), mat)))
```

```python
dist_mat
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
color = {0: 'r', 1: 'g'}
ax.scatter(mat[:, 0], mat[:, 1], mat[:, 2], c = list(map(lambda x : color[x], mat[:,3])))
ax.scatter(0,0,0)
```
