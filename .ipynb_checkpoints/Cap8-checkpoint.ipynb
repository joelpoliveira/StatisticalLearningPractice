## Random Forests / Bagging

```python
df = get_rdataset("Boston", "MASS", cache=True).data
```

```python
X = df.drop(columns="medv")
y = df.medv
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=2)
```

```python
bag = BaggingRegressor(n_estimators=500, random_state=1).fit(X_train, y_train)
print("score =", bag.score(X_train, y_train))

pred = bag.predict(X_train)
mse = ( (pred - y_train)**2 ).mean()
print("train mse =", mse)

pred = bag.predict(X_test)
mse = ( (pred - y_test)**2 ).mean()
print("test mse =", mse)
```

```python
rf = RandomForestRegressor(max_features=13, n_estimators=500, random_state=1).fit(X_train, y_train)
rf.score(X_train, y_train)

print("score =", rf.score(X_train, y_train))

pred = rf.predict(X_train)
mse = ( (pred - y_train)**2 ).mean()
print("train mse =", mse)

pred = rf.predict(X_test)
mse = ( ( pred - y_test)**2 ).mean()
print("test mse =", mse)
```

this random forest model uses all predictors / split, which is equivalent to the bagging. Therefore, the results are pretty similar with the bagging model

```python
rf = RandomForestRegressor(max_features=6, random_state=1).fit(X_train, y_train)
rf.score(X_train, y_train)

print("score =", rf.score(X_train, y_train))

pred = rf.predict(X_train)
mse = ( (pred - y_train)**2 ).mean()
print("train mse =", mse)

pred = rf.predict(X_test)
mse = ( ( pred - y_test)**2 ).mean()
print("test mse =", mse)
```

```python
n_feat_error = pd.Series(dtype=float)
n_tree_error = pd.Series(dtype=float)

for i in range(2,13+1):
    rf = RandomForestRegressor(max_features=i).fit(X_train, y_train)
    
    pred = rf.predict(X_test)
    n_feat_error.loc[i] = mean_squared_error(y_test, pred)

for i in range(1, 500, 20):
    rf = RandomForestRegressor(n_estimators=i).fit(X_train, y_train)
    
    pred = rf.predict(X_test)
    n_tree_error.loc[i] = mean_squared_error(y_test, pred)

f, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].plot(n_feat_error, "--*")
ax[0].set_xlabel("Number of features")
ax[0].set_ylabel("Test error")

ax[1].plot(n_tree_error, "--*")
ax[1].set_xlabel("Number of Trees")
```

Independently, raising the number of trees leads to a start improvement of the result, that remains stable for the more trees we add. Setting the number of features to consider reduces the value of the test error, only to a certain number of features. It seems that if we consider every feature, the error increases again ( which may be due to overfitting)

> 8.c)

```python
def mse(estimator, X, y):
    pred = estimator.predict(X)
    return ( (pred - y)**2 ).mean()
```

```python
def tree_prunning(tree, X_train, y_train, X_test, y_test):
    path = tree.cost_complexity_pruning_path(X_train, y_train)
    alphas, impurities = path.ccp_alphas, path.impurities
    
    scoring_board = pd.DataFrame(columns=["alpha", "mean", "std", "tree"])
    cv = KFold(10)
    for i, a in enumerate(alphas):
        model = DecisionTreeRegressor(ccp_alpha=a).fit(X_train, y_train)
        
        cv_err = cross_val_score(model, X_test, y_test, cv = cv, scoring=mse)
        scoring_board.loc[i] = {"alpha": a, 
                                "mean": cv_err.mean(), 
                                "std":cv_err.std(),
                               "tree":model}
    return scoring_board
```

```python
score_board = tree_prunning(dtr, X_train, y_train, X_test, y_test)
```

```python
score_board["mean"].argmin(), score_board["mean"].min()
```

```python
bag = BaggingRegressor().fit(X_train, y_train)

pred = bag.predict(X_test)
print(
    "MSE =", np.sqrt( mean_squared_error(y_test, pred)),
    "\nMean =", y.mean(),
    "\nStdDev =", y.std()
)
```
